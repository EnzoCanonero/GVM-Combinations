```python
# Define parameters
comb = 'lhc'
uncertainties = {}
input_folder = "input_files/"
output_folder = "output_files"

# Initialize combination object
top_comb = combination(input_folder, comb=comb)

# Add a measurement
syst_name = "new"
m_t_new = 174.5
stat_new = 0.4
syst_value = 0.5
uncertain_syst_value = 0.1
top_comb.add_measurement(m_t_new, stat_new, syst_name, syst_value, uncertain_syst_value)

# Define systematic uncertainties and parameters
systematics = ['new']
eps_end = 0.6
n_points = 7
ylim = (0., 0.5)

# Compute confidence intervals for orders 0 to 5
all_CIs = {}
for order in range(3):
    method = f'order_{order}'
    output_path = f'{output_folder}/CI_order_{order}'
    all_CIs[order] = top_comb.analyze_systematics_combination_CI(systematics, eps_end, n_points, ylim, method = method, output_path = output_path )

# Add numerical maximization
method_numerical = 'L-BFGS-B'
numerical_CI = top_comb.analyze_systematics_combination_CI(systematics, eps_end, n_points, ylim, method=method_numerical, output_path=f'{output_folder}/CI_numerical')

# Plotting
plt.figure(figsize=(11, 7))
epsilon_values = np.linspace(0.0, eps_end, n_points)

for order, CI in all_CIs.items():
    plt.plot(epsilon_values, CI[0], '.-', label=f'Order {order}', markersize=8)

# Plot numerical maximization
plt.plot(epsilon_values, numerical_CI[0], '.--', label='Numerical Maximization', color = 'black', markersize=8)

plt.xlabel('$_s$', fontsize=24)
plt.ylabel('$68.3\%$ Half-Size Confidence Interval (GeV)', fontsize=20)
plt.legend(fontsize=15, loc='lower left')
plt.grid(True)
plt.xlim(0, eps_end)
plt.ylim(*ylim)
plt.tight_layout()
plt.tick_params(axis='both', which='major', labelsize=16)
plt.savefig(f'{output_folder}/compare_CI_new_new1.pdf', bbox_inches="tight", format = 'pdf')
plt.show()
```

```python
input_folder = "input_files/"
output_folder = "output_files"

comb = 'lhc'
uncertainties = {}
top_comb = combination(input_folder, comb=comb)

syst_name = "new"
m_t_new = 174.5
stat_new = 0.2
syst_value = 0.25
uncertain_syst_value = 0.1
top_comb.add_measurement(m_t_new, stat_new, syst_name, syst_value, uncertain_syst_value)

# Define systematic uncertainties and parameters
systematics = ['CMSbHad']
eps_end = 0.6
n_points = 14
ylim = (172.4, 173.9)

# Compute central values for orders 0 to 5
all_CVs = {}
for order in range(0, 11, 2):
    method = f'order_{order}'
    output_path = f'{output_folder}/CV_order_{order}'
    all_CVs[order] = top_comb.analyze_systematics_combination_CV(systematics, eps_end=eps_end, n_points=n_points, ylim=ylim, method=method, output_path=output_path)

# Add numerical maximization
method_numerical = 'L-BFGS-B'
numerical_CV = top_comb.analyze_systematics_combination_CV(systematics, eps_end, n_points, ylim, method=method_numerical, output_path=f'{output_folder}/CV_numerical')

# Plotting
plt.figure(figsize=(11, 7))
epsilon_values = np.linspace(0.0, eps_end, n_points)

for order, CV in all_CVs.items():
    plt.plot(epsilon_values, CV, '.-', label=f'Order {order}', markersize=8)

# Plot numerical maximization
plt.plot(epsilon_values, numerical_CV, '.--', label='Numerical Maximization', color = 'black', markersize=8)

plt.xlabel('$_s$', fontsize=24)
plt.ylabel('Central Value (GeV)', fontsize=20)
plt.legend(fontsize=15, loc='lower left')
plt.grid(True)
plt.xlim(0, eps_end)
plt.ylim(*ylim)
plt.tight_layout()
plt.tick_params(axis='both', which='major', labelsize=16)
plt.savefig(f'{output_folder}/compare_CV_CMSbHad_new2.pdf', bbox_inches="tight", format = 'pdf')
plt.show()
```
